We can give the axes (x and y) titles in the graph 

```python
x = [1, 3, 4]
y = [10, 9, 8]
plt.plot(x, y)
plt.title('Just a graph!')
plt.xlabel('Horizontal Axis')
plt.ylabel('Vertical Axis')
plt.show()
```

Let's edit the font

```python
x = [1, 3, 4]
y = [10, 9, 8]
plt.plot(x, y)
plt.title('Just a graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': 30}) # size or fontname
plt.xlabel('Horizontal Axis', fontdict={'fontname': 'Impact', 'size': 15})
plt.ylabel('Vertical Axis', fontdict={'fontname': 'MS Gothic', 'fontsize': 40})
plt.show()
```

Now let's modify the ticks in the graph, in order to get rid of the decimals, this can be useful if you have to plot a graph using certain values in the axes

Add a legend to the line drawn

```python
x = [1, 3, 4]
y = [10, 9, 8]
plt.plot(x, y, label='Budget')
plt.title('Just a graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': 30}) # size or fontname
plt.xlabel('Horizontal Axis', fontdict={'fontname': 'Impact', 'size': 15})
plt.ylabel('Vertical Axis', fontdict={'fontname': 'MS Gothic', 'fontsize': 40})
plt.legend()
plt.show()
```

Changing colors

```python
x = [1, 3, 4]
y = [10, 9, 8]
plt.plot(x, y, label='Budget', color= 'green') # Name of the color
plt.title('Just a graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': 30}) # size or fontname
plt.xlabel('Horizontal Axis', fontdict={'fontname': 'Impact', 'size': 15})
plt.ylabel('Vertical Axis', fontdict={'fontname': 'MS Gothic', 'fontsize': 40})
plt.legend()
plt.show()
```

We can also define the color using its hexadecimal value

```python
x = [1, 3, 4]
y = [10, 9, 8]
plt.plot(x, y, label='Budget', color= '#ababab') # Hexadecimal color
plt.title('Just a graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': 30}) # size or fontname
plt.xlabel('Horizontal Axis', fontdict={'fontname': 'Impact', 'size': 15})
plt.ylabel('Vertical Axis', fontdict={'fontname': 'MS Gothic', 'fontsize': 40})
plt.legend()
plt.show()
```

Playing with the line width and datapoints of the line

Let's play with marker colors

Last cell has a sizeable amount of code, let's use the shorthand notation for these types of customizations

```python
x = [1, 3, 4]
y = [10, 9, 8]
# Normal notation
# plt.plot(x, y, label='Budget', color= '#ababab', linewidth=5, marker='.', linestyle='--', markersize=20, markeredgecolor='red') 
# Shorthand Notation plt.title('Just a graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': 30})
# fmt = '[first letter of color][marker][line]' 
plt.plot(x, y,'b.--', label='Just a graph')
plt.xlabel('Horizontal Axis', fontdict={'fontname': 'Impact', 'size': 15})
plt.ylabel('Vertical Axis', fontdict={'fontname': 'MS Gothic', 'fontsize': 40})
plt.legend()
plt.show()
```

Adding more lines in a single graph

```python
x = [1, 3, 4]
y = [10, 9, 8]
# Normal notation
# plt.plot(x, y, label='Budget', color= '#ababab', linewidth=5, marker='.', linestyle='--', markersize=20, markeredgecolor='red') 
# Shorthand Notation plt.title('Just a graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': 30})
# fmt = '[first letter of color][marker][line]' 
plt.plot(x, y,'b.--', label='Just a graph')
plt.xlabel('Horizontal Axis', fontdict={'fontname': 'Impact', 'size': 15})
plt.ylabel('Vertical Axis', fontdict={'fontname': 'MS Gothic', 'fontsize': 40})
# Create an array. 
line_2 = np.arange(0, 10, 2)
print(line_2)
plt.plot(line_2, line_2**2, 'k', label= 'line_2^2')
plt.legend()
plt.show()
```

Slicing or filtering a line

```python
x = [1, 3, 4]
y = [10, 9, 8]
plt.plot(x, y,'b.--', label='Just a graph')
plt.xlabel('Horizontal Axis', fontdict={'fontname': 'Impact', 'size': 15})
plt.ylabel('Vertical Axis', fontdict={'fontname': 'MS Gothic', 'fontsize': 40})
line_2 = np.arange(0, 10, 2)
print(line_2[0:3])
plt.plot(line_2[0:3], line_2[0:3]**2, 'k-*', label= 'line_2^2')
plt.plot()
plt.legend()
plt.show()
```

Applying a second format to a line 

```python
x = [1, 3, 4]
y = [10, 9, 8]
plt.plot(x, y,'b.--', label='Just a graph')
plt.xlabel('Horizontal Axis', fontdict={'fontname': 'Impact', 'size': 15})
plt.ylabel('Vertical Axis', fontdict={'fontname': 'MS Gothic', 'fontsize': 40})
line_2 = np.arange(0, 10, 2)
print(line_2)
plt.plot(line_2[0:3], line_2[0:3]**2, 'k-*', label= 'line_2^2')
plt.plot(line_2[2:], line_2[2:]**2, 'k--.') # we have to slice using the previous item
plt.plot()
plt.legend()
plt.show()
```

Resize the graph

```python
x = [1, 3, 4]
y = [10, 9, 8]
plt.figure(figsize=(6,5), dpi=300) # X to Y Ratio, DPI
plt.plot(x, y,'b.--', label='Just a graph')
plt.xlabel('Horizontal Axis', fontdict={'fontname': 'Impact', 'size': 15})
plt.ylabel('Vertical Axis', fontdict={'fontname': 'MS Gothic', 'fontsize': 40})
line_2 = np.arange(0, 10, 2)
print(line_2)
plt.plot(line_2[0:3], line_2[0:3]**2, 'k-*', label= 'line_2^2')
plt.plot(line_2[2:], line_2[2:]**2, 'k--.') # we have to slice using the previous item
plt.plot()
plt.legend()
plt.show()
```
